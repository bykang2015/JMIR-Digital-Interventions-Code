```python
from scipy.stats import shapiro

data = [2, 1, 2, 4, 2, 3, 4, 2, 1, 2, 2, 2, 3, 1, 1, 2, 2, 2, 4, 4, 3, 2, 4, 4, 2, 3, 1, 4]

stat, p_value = shapiro(data)

print(f'Shapiro-Wilk test statistic: {stat:.4f}, p-value: {p_value:.4f}')

if p_value > 0.05:
    print("The data appears to be normally distributed.")
else:
    print("The data does not appear to be normally distributed.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data
data = [2, 1, 2, 4, 2, 3, 4, 2, 1, 2, 2, 2, 3, 1, 1, 2, 2, 2, 4, 4, 3, 2, 4, 4, 2, 3, 1, 4]

# Create a Box-and-Whisker Plot with gradient
sns.boxplot(x=data, palette="viridis")

# Customize the plot
plt.title('Self Report Mental health literacy(5 scale)')
plt.xlabel('Mental Health Literacy Level')
plt.ylabel('Value')

# Show the plot
plt.show()
```
